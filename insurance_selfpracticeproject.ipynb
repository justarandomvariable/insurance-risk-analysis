### applying pca for dimensioanlity reduction

```python
pca_features=['gross_premium',
 'net_premium',
 'net_premium_earned',
 'incurred_claims_net',
 'commission_net',
 'management_expenses_m.e.',
 'invt._income_excluding_of_profit/loss_on_sale_of_invt.',
 'invt._income_including_of_p/l_on_sale_of_invt.',
 'total_investment_income',
 'profit_before_tax_pbt',
 'profit_after_tax_pat']
```

```python
from sklearn.decomposition import PCA
pca=PCA()
pca=PCA(n_components=2)
x_pca=pca.fit_transform(x_scaled_df[pca_features])
```

```python











x_pca_df=pd.DataFrame(x_pca,
                     columns=['pca_1','pca_2'],
                     index=df.index)
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
import pickle

with open("insurance_final_data.pkl", "wb") as f:
    pickle.dump(
        {
            "X_scaled": x_scaled_df,   # for clustering / interpretation
            "X_pca": x_pca_df,         # for supervised models
            "scaler": scaler,
            "pca": pca
        },
        f
    )
```

```python
import os
os.path.exists("insurance_final_data.pkl")
```

```python
from sklearn.model_selection import train_test_split
X=x_scaled_df
Y=df['claims_ratio']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print("R2:", r2_score(Y_test,y_pred))
print("MSE:", mean_squared_error(Y_test,y_pred))
```

```python
print("Train_r2:", r2_score(Y_train,(model.predict(X_train))))
print("Test_r2:",r2_score(Y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_squared_error
ridge=Ridge(alpha=0.1)
ridge.fit(X_train,Y_train)
```

```python
y_pred_ridge=ridge.predict(X_test)
```

```python
print("Ridge_r2:", r2_score(Y_test,y_pred_ridge))
print("Ridge_mse:", mean_squared_error(Y_test,y_pred_ridge))
```

```python
print("Train_R2_Ridge:", r2_score(Y_train, ridge.predict(X_train)))
print("Test_R2_Ridge:", r2_score(Y_test,y_pred_ridge))
```

```python
import pickle

with open ("insurance_ridge_model.pkl", "wb") as f:
    pickle.dump(ridge,f)
```
